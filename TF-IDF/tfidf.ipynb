```python
import pandas as pd
import numpy as np
df = pd.read_csv('../data.csv')
```

```python
df.head(1)
```

```python
import jieba
df['cut'] = df['text'].apply(lambda x: [i for i in jieba.cut(x)]) #分词
```

```python
corpus = []
for i in df['cut'].values:
    corpus += i
```

```python
corpus = list(set(corpus)) #词表
```

词频（TF） = 某个词在文章中的出现次数

```python
tf = np.zeros(((len(df)), len(corpus))) #row：每句话；col：每个词

for row, sentence in enumerate(df['cut']):
    for col, word in enumerate(corpus):
        tf[row][col] = sentence.count(word) #直接统计出现次数 未进行归一化
    
    
```

逆文档频率（IDF） = log（语料库的文档总数/包含该词的文档总数）

```python
idf = np.zeros((len(corpus), len(df))) #实际上idf为len(corpus)维度，考虑到tfidf计算，将其进行broadcast

for row, word in enumerate(corpus):
    df_ = 0
    for col, sentence in enumerate(df['cut']):
        if word in sentence:
            df_ += 1
    idf[row] = np.log(len(df)/ df_)
    
```

```python
tf_idf = tf * idf.T
```

```python
tf_idf.shape
```

```python
pd.DataFrame(tf, columns=corpus)
```

```python
pd.DataFrame(idf.T, columns=corpus)
```

```python
pd.DataFrame(tf_idf, columns=corpus)
```
